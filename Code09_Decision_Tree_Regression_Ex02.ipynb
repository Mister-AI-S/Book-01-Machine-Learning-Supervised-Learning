## Step 1: Import Necessary Libraries

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

## Step 2: Create the Dataset and plot the data

#### We use the same dataset for this example:

```python
X = np.array([[0.5, 1.1, 1.6, 2.2, 2.8, 3.3, 3.7, 3.9, 4.5, 4.9, 5.3, 5.8, 6.5, 6.9, 7.4, 7.8]]).reshape(-1,1)

y = np.array([2.3, 2.4, 2.3, 2.4, 2.3, 0.5, 0.6, 0.49, 3.4, 3.5, 3.4, 3.4, 2.7, 2.9, 2.8, 3])

# Plotting the data
plt.scatter(X, y, color='blue')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Data")
plt.show()
```

## Step 3: Train a Decision Tree Regressor
#### Create an instance of the Decision Tree Regressor.
#### Train the model on the dataset.

```python
model = DecisionTreeRegressor(max_depth=2)
model.fit(X, y)
```

## Step 4: Make Predictions

#### We can now predict target for new samples.

```python

sample = np.array([[1.3]])
prediction = model.predict(sample)

print(f"Predicted target: {prediction[0]:.2f}")
```

## Step 5: Visualize the Decision Tree

#### Visualize the tree structure to better understand how the model splits the data.

```python
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=['Feature'], filled=True)
plt.show()
```
